```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('../../Data/modelTraining/trainData.csv')
test_df = pd.read_csv('../../Data/modelTraining/testData.csv')
cleaned_data = pd.read_csv('../../Data/cleanedData/CleanedScaledData.csv')
```

```python
train_df_annotated = cleaned_data[cleaned_data['Target Gene Symbol'].isin(train_df['Target_Gene_Symbol'])]
test_df_annotated = cleaned_data[cleaned_data['Target Gene Symbol'].isin(test_df['Target_Gene_Symbol'])]
train_df_annotated
```

```python
train_df_unique = train_df_annotated.groupby(['Target Gene Symbol','sgRNA \'Cut\' Site TSS Offset','chromosome','Strand of Target',
                                              'sgRNA \'Cut\' Position','sgRNA Context Sequence','Orientation','rs3ChenSeqScore',
                                              'ATAC overlap with peak proportion'])['doubleZscore'].mean().reset_index()
test_df_unique = test_df_annotated.groupby(['Target Gene Symbol','sgRNA \'Cut\' Site TSS Offset','chromosome','Strand of Target',
                                              'sgRNA \'Cut\' Position','sgRNA Context Sequence','Orientation','rs3ChenSeqScore',
                                              'ATAC overlap with peak proportion'])['doubleZscore'].mean().reset_index()
train_df_unique
```

```python
train_df_unique['chromosome_crispick_input'] = train_df_unique['chromosome']
train_df_unique['chromosome_crispick_input'] = train_df_unique['chromosome_crispick_input'].replace('chr23','chrX')
train_df_unique['chromosome_crispick_input'] = train_df_unique['chromosome_crispick_input'].replace('chr24','chrY')

coordinates = (train_df_unique['chromosome_crispick_input'] + ':' +
              train_df_unique['Strand of Target'] + ':' +
              train_df_unique['sgRNA \'Cut\' Position'].astype(int).astype(str))
coordinates
```

```python
train_df_unique['coordinates'] = coordinates
coordinates.drop_duplicates()[0:50000].to_csv('../../Data/modelTraining/rs3i_training_coordinates_part1.txt',
                                     sep = '\n',header=None,index=False)
coordinates.drop_duplicates()[50000:].to_csv('../../Data/modelTraining/rs3i_training_coordinates_part2.txt',
                                     sep = '\n',header=None,index=False)
```

```python
test_df_unique['chromosome_crispick_input'] = test_df_unique['chromosome']
test_df_unique['chromosome_crispick_input'] = test_df_unique['chromosome_crispick_input'].replace('chr23','chrX')
test_df_unique['chromosome_crispick_input'] = test_df_unique['chromosome_crispick_input'].replace('chr24','chrY')

coordinates = (test_df_unique['chromosome_crispick_input'] + ':' +
              test_df_unique['Strand of Target'] + ':' +
              test_df_unique['sgRNA \'Cut\' Position'].astype(int).astype(str))
coordinates
```

```python
test_df_unique['coordinates'] = coordinates
coordinates.drop_duplicates().to_csv('../../Data/modelTraining/rs3i_testing_coordinates.txt',
                                     sep = '\n',header=None,index=False)
```

#### Run CRISPick with following inputs:

Reference genome: Ensembl v.114\
Mechanism: CRISPRi\
Enzyme: SpyoCas9\
On-Target Scorer: RS3i-Chen2013+ATAC\
Off-Target Scorer: Seed Score v1\
Targets: upload file with coordinates\
Quota: 1\
Click report unpicked sequences\
TSS range: [-20,20]

```python
crispick_train_part1 = pd.read_table('../../Data/modelTraining/rs3i_training_designs_part1.txt')
crispick_train_part2 = pd.read_table('../../Data/modelTraining/rs3i_training_designs_part2.txt')
crispick_test = pd.read_table('../../Data/modelTraining/rs3i_testing_designs.txt')
crispick = pd.concat([crispick_part1,crispick_part2,crispick_test])
```

```python
#left merge to see if there are any training data that do not get an atac-seq score
train_df_with_atac = train_df_unique.merge(crispick[['Input','sgRNA Context Sequence','ATAC-seq Score']].drop_duplicates(),
                                        left_on = ['coordinates','sgRNA Context Sequence'],
                                        right_on = ['Input','sgRNA Context Sequence'],
                                        how = 'left')
test_df_with_atac = test_df_unique.merge(crispick[['Input','sgRNA Context Sequence','ATAC-seq Score']].drop_duplicates(),
                                        left_on = ['coordinates','sgRNA Context Sequence'],
                                        right_on = ['Input','sgRNA Context Sequence'],
                                        how = 'left')
```

```python
#check that all guides got an atac-seq score
train_df_with_atac['ATAC-seq Score'].isna().sum()
```

```python
#see that most of atac scores agree
(train_df_with_atac['ATAC-seq Score'] == round(train_df_with_atac['ATAC overlap with peak proportion'],4)).mean()
```

```python
(test_df_with_atac['ATAC-seq Score'] == round(test_df_with_atac['ATAC overlap with peak proportion'],4)).mean()
```

```python
def clean_column_name(name):
    return (
        str(name)
        .replace('[', '_')
        .replace(']', '_')
        .replace('<', '_')
        .replace('>', '_')
        .replace(' ', '_')
        .replace(',', '_')
        .replace('(', '_')
        .replace(')', '_')
    )
```

```python
train_df_with_atac.columns = [clean_column_name(col) for col in train_df_with_atac.columns]
test_df_with_atac.columns = [clean_column_name(col) for col in test_df_with_atac.columns]
```

```python
X_train = train_df_with_atac[['Target_Gene_Symbol','sgRNA_Context_Sequence','sgRNA_\'Cut\'_Site_TSS_Offset', 'rs3ChenSeqScore','doubleZscore']]
X_train.loc[:,'rs3ChenSeqScore'] = round(X_train['rs3ChenSeqScore'],4)
X_train.loc[:,'sgRNA_\'Cut\'_Site_TSS_Offset'] = X_train['sgRNA_\'Cut\'_Site_TSS_Offset'].astype(int)
X_train['ATAC_overlap_with_peak'] = pd.cut(train_df_with_atac['ATAC-seq_Score'],bins=[0,0.5,1,np.inf],right=False)

# one-hot encoding
X_train_encoded = pd.get_dummies(X_train, columns=['ATAC_overlap_with_peak'])

bool_cols = X_train_encoded.select_dtypes(include=['bool']).columns.tolist()
X_train_encoded[bool_cols] = X_train_encoded[bool_cols].astype(int)

def clean_column_name(name):
    return (
        str(name)
        .replace('[', '_')
        .replace(']', '_')
        .replace('<', '_')
        .replace('>', '_')
        .replace(' ', '_')
        .replace(',', '_')
        .replace('(', '_')
        .replace(')', '_')
    )

X_train_encoded.columns = [clean_column_name(col) for col in X_train_encoded.columns]
X_train_encoded
```

```python
X_test = test_df_with_atac[['Target_Gene_Symbol','sgRNA_Context_Sequence','sgRNA_\'Cut\'_Site_TSS_Offset', 'rs3ChenSeqScore','doubleZscore']]
X_test.loc[:,'rs3ChenSeqScore'] = round(X_test['rs3ChenSeqScore'],4)
X_test.loc[:,'sgRNA_\'Cut\'_Site_TSS_Offset'] = X_test['sgRNA_\'Cut\'_Site_TSS_Offset'].astype(int)
X_test['ATAC_overlap_with_peak'] = pd.cut(test_df_with_atac['ATAC-seq_Score'],bins=[0,0.5,1,np.inf],right=False)

# one-hot encoding
X_test_encoded = pd.get_dummies(X_test, columns=['ATAC_overlap_with_peak'])

bool_cols = X_test_encoded.select_dtypes(include=['bool']).columns.tolist()
X_test_encoded[bool_cols] = X_test_encoded[bool_cols].astype(int)

def clean_column_name(name):
    return (
        str(name)
        .replace('[', '_')
        .replace(']', '_')
        .replace('<', '_')
        .replace('>', '_')
        .replace(' ', '_')
        .replace(',', '_')
        .replace('(', '_')
        .replace(')', '_')
    )

X_test_encoded.columns = [clean_column_name(col) for col in X_test_encoded.columns]
X_test_encoded
```

```python
X_train_encoded.to_csv('../../Data/modelTraining/trainData_rs3i.csv',index=False)
X_test_encoded.to_csv('../../Data/modelTraining/testData_rs3i.csv',index=False)
```
