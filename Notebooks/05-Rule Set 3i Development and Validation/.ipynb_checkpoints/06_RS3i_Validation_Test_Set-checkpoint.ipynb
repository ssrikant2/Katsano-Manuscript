```python
import xgboost as xgb
import pandas as pd
import numpy as np
from scipy.stats import spearmanr
import matplotlib.pyplot as plt
import gpplot as gpp
import seaborn as sns
from statannotations.Annotator import Annotator
from sklearn.metrics import f1_score
gpp.set_aesthetics(context='paper')
```

```python
#load and format testing data for full-featured model
modelTrainingPath = "../../Data/modelTraining/"
test_df = pd.read_csv(f'{modelTrainingPath}testData.csv')
Xtest_full = test_df.drop(columns=['doubleZscore', 'Target_Gene_Symbol','sgRNA_Context_Sequence']).copy()
Xtest_full
```

```python
#format for RS3i
Xtest_rs3i = test_df[['sgRNA_\'Cut\'_Site_TSS_Offset', 'rs3ChenSeqScore']].copy()
Xtest_rs3i.loc[:,'rs3ChenSeqScore'] = round(Xtest_rs3i['rs3ChenSeqScore'],4)
Xtest_rs3i.loc[:,'sgRNA_\'Cut\'_Site_TSS_Offset'] = Xtest_rs3i['sgRNA_\'Cut\'_Site_TSS_Offset'].copy().astype(int)
Xtest_rs3i['ATAC_overlap_with_peak'] = pd.cut(test_df['ATAC_overlap_with_peak_proportion'],bins=[0,0.5,1,np.inf],right=False)

# one-hot encoding
Xtest_rs3i_encoded = pd.get_dummies(Xtest_rs3i, columns=['ATAC_overlap_with_peak'])

bool_cols = Xtest_rs3i_encoded.select_dtypes(include=['bool']).columns.tolist()
Xtest_rs3i_encoded[bool_cols] = Xtest_rs3i_encoded[bool_cols].astype(int)

def clean_column_name(name):
    return (
        str(name)
        .replace('[', '_')
        .replace(']', '_')
        .replace('<', '_')
        .replace('>', '_')
        .replace(' ', '_')
        .replace(',', '_')
        .replace('(', '_')
        .replace(')', '_')
    )

Xtest_rs3i_encoded.columns = [clean_column_name(col) for col in Xtest_rs3i_encoded.columns]
Xtest_rs3i_encoded
```

```python
Xtest_rs3i_without_atac = Xtest_rs3i[['sgRNA_\'Cut\'_Site_TSS_Offset','rs3ChenSeqScore']].copy()
```

```python
#load models
full_featured_model = xgb.XGBRegressor()
full_featured_model.load_model('../../Data/modelTraining/full_featured_model.json')

rs3i = xgb.XGBRegressor()
rs3i.load_model('../../Data/modelTraining/RS3i.json')

rs3i_without_atac = xgb.XGBRegressor()
rs3i_without_atac.load_model('../../Data/modelTraining/RS3i_without_ATAC.json')
```

```python
test_df['full_featured'] = full_featured_model.predict(Xtest_full)
test_df['RS3i'] = rs3i.predict(Xtest_rs3i_encoded)
test_df['RS3i_without_ATAC'] = rs3i_without_atac.predict(Xtest_rs3i_without_atac)
test_df
```

### Comparing performance across three models

```python
#minimum guides per gene?
test_df.groupby('Target_Gene_Symbol')['sgRNA_Context_Sequence'].nunique().min()
```

```python
def calculate_spearman_within_groups(df,group_col,y_col):
    spearman_corrs = []
    groups = df[group_col].unique()
    
    for group in groups:
        group_data = df[df[group_col] == group]
        full_corr, _ = spearmanr(group_data[y_col], group_data['full_featured'])
        group_data = group_data[['sgRNA_Context_Sequence',y_col,'RS3i','RS3i_without_ATAC']].groupby('sgRNA_Context_Sequence').mean().reset_index()
        rs3i_corr, _ = spearmanr(group_data[y_col], group_data['RS3i'])
        rs3i_no_ATAC_corr, _ = spearmanr(group_data[y_col], group_data['RS3i_without_ATAC'])
        spearman_corrs.append({group_col: group,
                               'Full-Featured Model': full_corr,
                               'RS3i': rs3i_corr,
                               'RS3i Without ATAC':rs3i_no_ATAC_corr})
    
    return pd.DataFrame(spearman_corrs)
```

```python
spearman_df = calculate_spearman_within_groups(test_df,'Target_Gene_Symbol','doubleZscore')
```

```python
plot_df = spearman_df.melt(id_vars = 'Target_Gene_Symbol', var_name = 'Model', value_name = 'spearmanr')
sns.boxplot(data = plot_df, x = 'Model', y = 'spearmanr', color = '#008BFB')
plt.ylabel('Spearman Correlation')
plt.title('Within-gene Spearman Correlation on Test Set (n = 94)')
sns.despine()

gpp.savefig('../../Figures/Figure 4/model_comparison_spearmanr_boxplot.pdf', dpi=300, bbox_inches = 'tight')
```

```python
#merge with dataframe labeled with original dataset(s) that each guide/gene was present in
labeled_data = pd.read_csv('../../Data/cleanedData/CombinedCleanedDatasets.csv')
labeled_data = labeled_data.groupby(['sgRNA Context Sequence','DataSet'])['signed_zscore'].mean().reset_index()
labeled_data
```

```python
test_df_with_dataset = test_df.merge(labeled_data, left_on = 'sgRNA_Context_Sequence', right_on = 'sgRNA Context Sequence', how = 'inner')
test_df_with_dataset['gene_dataset'] = test_df_with_dataset['Target_Gene_Symbol'] + '_' + test_df_with_dataset['DataSet']
test_df_with_dataset
```

```python
#minimum guides per gene?
test_df_with_dataset.groupby('gene_dataset')['sgRNA_Context_Sequence'].nunique().min()
```

```python
spearman_df = calculate_spearman_within_groups(test_df_with_dataset,'gene_dataset','signed_zscore')
spearman_df[['Target_Gene_Symbol','Original Dataset']] = spearman_df['gene_dataset'].str.split('_',expand=True)
spearman_df['Original Dataset'] = spearman_df['Original Dataset'].replace('InHouse','This Study')
spearman_df
```

### Comparing RS3i performance on train vs test sets

```python
#load and format training/testing data
train_df = pd.read_csv(f'{modelTrainingPath}trainData_rs3i.csv')
Xtrain = train_df.drop(columns=['doubleZscore', 'Target_Gene_Symbol','sgRNA_Context_Sequence']).copy()
Xtrain
```

```python
test_df = pd.read_csv(f'{modelTrainingPath}testData_rs3i.csv')
Xtest = test_df.drop(columns=['doubleZscore', 'Target_Gene_Symbol','sgRNA_Context_Sequence']).copy()
Xtest
```

```python
train_df['RS3i'] = rs3i.predict(Xtrain)
test_df['RS3i'] = rs3i.predict(Xtest)
```

```python
train_df['split'] = 'train'
test_df['split'] = 'test'
```

```python
train_test_df = pd.concat([train_df,test_df])
spearman_df = train_test_df.groupby(['Target_Gene_Symbol',
                                     'split']).apply(lambda x: spearmanr(x.doubleZscore,
                                                                         x.RS3i).statistic).reset_index().rename(columns = {0:'spearmanr'})
spearman_df
```

```python
fig,ax = plt.subplots(figsize = (5,5))
sns.boxplot(ax = ax, data = spearman_df, x = 'split', y = 'spearmanr')
plt.ylabel('Within-gene Spearman Correlation')
plt.title('RS3i Training and Testing Performance')
sns.despine()

pairs = [('train','test')]

annotator = Annotator(ax, pairs, data = spearman_df, x = 'split', y = 'spearmanr')
annotator.configure(test='Mann-Whitney', text_format='star', loc='inside')
annotator.apply_and_annotate()

gpp.savefig('../../Figures/Figure 4/rs3i_train_vs_test_spearmanr_boxplot.pdf', dpi=300, bbox_inches = 'tight')
```

### Defining a "good" RS3i Score

```python
test_df['RS3i'].hist(bins=50)
```

```python
labeled_data = labeled_data.groupby('sgRNA Context Sequence')['signed_zscore'].mean()
```

```python
#merge with z-scores to use to define "active" guides since they are directly interpretable as guide abundance relative to controls
test_df_with_zscores = test_df.merge(labeled_data, left_on = 'sgRNA_Context_Sequence', right_on = 'sgRNA Context Sequence', how = 'inner')
test_df_with_zscores
```

```python
test_df_with_zscores['signed_zscore'].hist(bins=50)
```

```python
#define bins
rs3iScore_bins = [-np.inf, 0, 0.4, 0.8, 1.2, np.inf]
zscore_bins = [-np.inf,0,1,2,3,np.inf]

#average z-scores for each unique context sequence (rs3i scores should be the same so averaging has no impact)
plot_df = test_df_with_zscores[['RS3i','signed_zscore','sgRNA_Context_Sequence']].groupby('sgRNA_Context_Sequence').mean().reset_index()

#bin rs3i scores and z-scores
plot_df['RS3i Score'] = pd.cut(plot_df['RS3i'],bins = rs3iScore_bins)
plot_df['Signed Z-Score'] = pd.cut(plot_df['signed_zscore'],bins = zscore_bins)

#calculate proportion of guides in each bin combination
plot_df = plot_df.groupby(['RS3i Score','Signed Z-Score']).agg(count = ('signed_zscore','nunique')).reset_index()
plot_df['total'] = plot_df.groupby(['RS3i Score'])['count'].transform('sum')
plot_df['proportion'] = plot_df['count']/plot_df['total']

#save total guides in each bin, sort beforehand to make sure list is ordered by RS3i score bin ascending
total_guides_in_bins = plot_df.sort_values(by = 'RS3i Score')['total'].unique().tolist()

#pivot df
plot_df = plot_df.pivot(columns = 'Signed Z-Score', index = 'RS3i Score', values = 'proportion')
plot_df
```

```python
plot_df.plot(kind='bar', stacked=True, edgecolor="none", cmap="Blues")
plt.xlabel('RS3i Score',fontsize=14)
plt.ylabel('Proportion of Guides',fontsize=14)
plt.title('Held-out Test Set (94 Genes)',fontsize=16)
plt.legend(title="Signed Z-Score", bbox_to_anchor=[1, 1], fontsize = 14)
sns.despine()

for i,n_guides in enumerate(total_guides_in_bins):
    plt.text(i,1,str(n_guides),ha='center',va='bottom',fontsize=12)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

gpp.savefig('../../Figures/Figure 4/rs3i_test_set_stacked_barplot.pdf', dpi=300, bbox_inches = 'tight')
```
