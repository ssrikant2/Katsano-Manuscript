### Transcript Selection

```python
#read in transcript annotations
protein_coding_transcripts = pd.read_csv('../../Data/GENCODE48/gencode_48_protein_coding_transcripts.csv')
protein_coding_transcripts
```

```python
#select Ensembl Canonical transcripts
ensembl_canonical_transcripts = protein_coding_transcripts[protein_coding_transcripts['tag'].str.contains('Ensembl_canonical')].copy()
ensembl_canonical_transcripts
```

```python
#read in illumina-selected transcripts, reformat to be compatible with gencode file
illumina_tss_df = pd.read_table('../../Data/Katsano/CAGE_tss.tsv')
illumina_tss_df['transcript_id_version'] = illumina_tss_df['transcript_id']
illumina_tss_df = illumina_tss_df.drop('transcript_id',axis=1)
illumina_tss_df['transcript_id_without_version'] = illumina_tss_df['transcript_id_version'].str.split('.', expand=True)[0]
illumina_tss_df['strand'] = np.where(illumina_tss_df['strand'] == 1, '+', '-')
illumina_tss_df
```

```python
#select illumina transcripts still present in GENCODE v48
illumina_transcripts = protein_coding_transcripts[protein_coding_transcripts['ID'].isin(illumina_tss_df['transcript_id_version'])].copy()
noncanonical_illumina = illumina_transcripts[~illumina_transcripts['ID'].isin(ensembl_canonical_transcripts['ID'])].copy()
```

```python
#merge illumina df with gencode protein-coding df on transcript id without version
merged_df = protein_coding_transcripts.merge(illumina_tss_df,
                                             left_on = ['seqid','strand','transcript_id_without_version'],
                                             right_on = ['chrom','strand','transcript_id_without_version'],
                                             how = 'inner')

merged_df = merged_df.rename(columns = {'gene_id_y':'gene_id_illumina','gene_id_x':'gene_id_gencode48',
                                        'transcript_id_version':'transcript_id_version_illumina',
                                        'ID':'transcript_id_version_gencode48',
                                        'gene':'gene_symbol_illumina','gene_name':'gene_symbol_gencode48',
                                        'TSS':'tss_gencode48'})
merged_df
```

```python
#illumina tss is one off from gencode
(merged_df['tss_pos'] - merged_df['tss_gencode48']).value_counts()
```

```python
#shift illumina tss to match gencode
merged_df['tss_illumina'] = merged_df['tss_pos'] + 1
```

```python
#find transcripts that have been updated
updated_transcripts = merged_df[merged_df['transcript_id_version_illumina'] != merged_df['transcript_id_version_gencode48']].copy()

#subset to those with unchanged tss
same_tss_updated_transcripts = updated_transcripts[updated_transcripts['tss_illumina'] == updated_transcripts['tss_gencode48']].copy()

#get the same transcript rows in the original dataframe so the columns match already selected transcripts
same_tss_updated_transcripts = protein_coding_transcripts[protein_coding_transcripts['ID'].isin(same_tss_updated_transcripts['transcript_id_version_gencode48'])].copy()

same_tss_updated_transcripts
```

```python
#subset to illumina-selected transcripts which aren't in gencode v48 and also haven't been selected already
illumina_not_in_gencode48 = illumina_tss_df[~illumina_tss_df['transcript_id_version'].isin(protein_coding_transcripts['ID']) &
                                            ~illumina_tss_df['transcript_id_without_version'].isin(same_tss_updated_transcripts['transcript_id_without_version'])]

#merge on chromosome and strand only to get all other transcripts on same chromosome+strand that are in gencode v48
chrom_strand_merged_df = protein_coding_transcripts.merge(illumina_not_in_gencode48,
                                                          left_on = ['seqid','strand'],
                                                          right_on = ['chrom','strand'],
                                                          how = 'inner')

chrom_strand_merged_df = chrom_strand_merged_df.rename(columns = {'gene_id_y':'gene_id_illumina',
                                                                  'gene_id_x':'gene_id_gencode48',
                                                                  'transcript_id_version':'transcript_id_version_illumina',
                                                                  'ID':'transcript_id_version_gencode48',
                                                                  'gene':'gene_symbol_illumina','gene_name':'gene_symbol_gencode48',
                                                                  'TSS':'tss_gencode48',
                                                                  'transcript_id_without_version_x':'transcript_id_gencode48',
                                                                  'transcript_id_without_version_y':'transcript_id_illumina'})

chrom_strand_merged_df['tss_illumina'] = chrom_strand_merged_df['tss_pos'] + 1

chrom_strand_merged_df
```

```python
#get one gencode v48 transcript per illumina transcript + gencode v48 gene with closest TSS
chrom_strand_merged_df['tss_diff'] = np.abs(chrom_strand_merged_df['tss_gencode48'] - chrom_strand_merged_df['tss_illumina'])
closest_transcript_idxs = chrom_strand_merged_df.groupby(['transcript_id_version_illumina','gene_id_gencode48'])['tss_diff'].idxmin()
closest_transcripts = chrom_strand_merged_df.loc[closest_transcript_idxs].copy()

#label whether the closest transcript for the given gene matches the gene/transcript id corresponding to the illumina transcript
closest_transcripts['same_gene_symbol'] = (closest_transcripts['gene_symbol_illumina'] == closest_transcripts['gene_symbol_gencode48'])
closest_transcripts['same_gene_id'] = (closest_transcripts['gene_id_illumina'] == closest_transcripts['gene_id_gencode48'])
closest_transcripts['same_transcript_id'] = (closest_transcripts['transcript_id_illumina'] == closest_transcripts['transcript_id_gencode48'])
closest_transcripts
```

```python
#subset to those associated with the same gene; allow for the same gene symbol, gene id, or transcript id since nomenclature has changed over time
same_gene_closest_transcripts = closest_transcripts[closest_transcripts['same_gene_symbol'] | closest_transcripts['same_gene_id'] |
                                                    closest_transcripts['same_transcript_id']].copy()

same_gene_closest_transcripts = protein_coding_transcripts[protein_coding_transcripts['ID'].isin(same_gene_closest_transcripts['transcript_id_version_gencode48'])].copy()

same_gene_closest_transcripts
```

Since original intergenic controls don't follow the same seed score distribution as targeting guides, resample them.

```python
onesite_controls = pd.read_table('../../Data/CRISPickRuns/sgrna-negcontrols-onesite-9606-GRCh38-SpyoCas9-10000.txt',skiprows=1).rename(columns = {'sgRNA Seq':'sgRNA Sequence'})[['sgRNA Sequence']]
```

```python
def count_overlapping_matches(seq,pattern):
    count = 0
    length = len(pattern)
    for i in range(len(seq)-1):
        if seq[i:(i+length)] == pattern:
            count +=1
    return count
```

```python
#calculate seed scores for all controls to be sampled from
onesite_controls['seed_seq'] = onesite_controls['sgRNA Sequence'].str[-12:]
onesite_controls['Seed Score (v1)'] = onesite_controls['seed_seq'].apply(lambda seq: count_overlapping_matches(seq,'GG'))
```

```python
num_onesite = 900

#get list of all targeting guide seed scores
targeting_seed_scores = katsano_picked.loc[katsano_picked['guide_group'] == 'targeting',
                                          ['sgRNA Sequence','Seed Score (v1)']].drop_duplicates()

#randomly sample number of intended onesite guides from targeting guides
sampled_seed_scores_num_onesite = random.sample(targeting_seed_scores['Seed Score (v1)'].tolist(),num_onesite)

selected_onesite_list = []

#loop through unique seed score values of targeting guides
unique_seed_scores = targeting_seed_scores['Seed Score (v1)'].unique().tolist()
for ss in unique_seed_scores:
    #subset controls to those with that particular seed score
    onesite_subset = onesite_controls.loc[onesite_controls['Seed Score (v1)'] == ss,'sgRNA Sequence'].tolist()

    #get the number of sampled targeting guides with that seed score
    num_sampled_onesite = sum([x==ss for x in sampled_seed_scores_num_onesite])

    #sample that number of guides from the control subsets
    selected_onesites = random.sample(onesite_subset,num_sampled_onesite)

    #add sampled controls to list
    selected_onesite_list = selected_onesite_list + selected_onesites
```

```python
#include original nosites
katsano_nosites = katsano_picked.loc[katsano_picked['guide_group'] == 'non-targeting',['sgRNA Sequence']].reset_index(drop=True)
katsano_nosites['guide_group'] = 'non-targeting'

#format onesites into df
katsano_onesites = pd.DataFrame({'sgRNA Sequence':selected_onesite_list})
katsano_onesites['guide_group'] = 'intergenic'

katsano_controls = pd.concat([katsano_nosites,katsano_onesites])
katsano_controls['seed_seq'] = katsano_controls['sgRNA Sequence'].str[-12:]
katsano_controls['Seed Score (v1)'] = katsano_controls['seed_seq'].apply(lambda seq: count_overlapping_matches(seq,'GG'))

#subset df of targeting
katsano_targeting = katsano_picked[~katsano_picked['Picking Round'].isna()]
```

```python
#katsano_controls[['sgRNA Sequence','guide_group']].to_csv('../../Data/Katsano/katsano_controls.txt',index=False,sep='\t')
```

```python
katsano_controls = pd.read_table('../../Data/Katsano/katsano_controls.txt')
katsano_controls['seed_seq'] = katsano_controls['sgRNA Sequence'].str[-12:]
katsano_controls['Seed Score (v1)'] = katsano_controls['seed_seq'].apply(lambda seq: count_overlapping_matches(seq,'GG'))

new_katsano = pd.concat([katsano_targeting,katsano_controls]).reset_index(drop=True)

plot_df = new_katsano.groupby(['guide_group','Seed Score (v1)']).agg(num_guides = ('sgRNA Sequence','nunique')).reset_index()

plot_df['total_guides'] = plot_df.groupby('guide_group')['num_guides'].transform('sum')
plot_df['prop_guides'] = plot_df['num_guides']/plot_df['total_guides']
plot_df = plot_df.pivot(index = 'guide_group', columns = 'Seed Score (v1)', values = 'prop_guides')

#remove non-targeting
plot_df = plot_df[plot_df.index != 'non-targeting'].copy()

plot_df.plot(kind='bar', stacked=True, cmap = 'Reds')
plt.legend(title = 'Seed Score', bbox_to_anchor = [1,1])
plt.ylabel('Proportion of guides')
plt.xlabel('Guide Type')
plt.xticks(rotation=0)
plt.title('Katsano Guide Seed Scores')

sns.despine()
gpp.savefig('../../Figures/Figure 7/katsano_targeting_vs_control_seed_scores_resampled_onesites.pdf',bbox_inches='tight',dpi=300)
```
