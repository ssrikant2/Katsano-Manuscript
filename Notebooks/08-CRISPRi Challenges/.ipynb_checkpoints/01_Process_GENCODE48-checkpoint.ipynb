```python
import pandas as pd
import numpy as np
from upsetplot import plot
import matplotlib.pyplot as plt
import gpplot as gpp
import seaborn as sns
gpp.set_aesthetics(context = 'paper')
```

```python
#load GENCODE v48 primary assembly GFF3 file
colnames = ['seqid', 'source', 'type', 'start', 'end','score', 'strand', 'phase', 'attributes']
gff_df = pd.read_csv('../../Data/GENCODE48/gencode.v48.primary_assembly.annotation.gff3', sep="\t", comment="#", names=colnames)

gff_df
```

```python
#subset to transcript info
transcript_df = gff_df[gff_df.type == 'transcript'].copy().reset_index(drop=True)
transcript_df
```

```python
#create columns with each attribute
expanded_attributes = transcript_df.attributes.str.split(';',expand=True).stack().str.split('=',expand=True).reset_index()
expanded_attributes_pivoted = expanded_attributes.pivot(columns = 0, values = 1, index = 'level_0').reset_index(drop=True)
expanded_attributes_pivoted
```

```python
#combine transcript info with newly expanded df of attributes
transcript_attribute_df = pd.concat([transcript_df,expanded_attributes_pivoted],axis=1)
transcript_attribute_df
```

```python
#some transcripts have 0 tags; replace NAs with ''
transcript_attribute_df['tag'] = transcript_attribute_df['tag'].replace(np.nan,'no_tags')
transcript_attribute_df['tag'].isna().sum()
```

```python
#subset to transcripts of protein-coding genes
protein_coding_transcripts = transcript_attribute_df[transcript_attribute_df['gene_type'] == 'protein_coding'].copy()
protein_coding_transcripts
```

```python
#compute TSS
protein_coding_transcripts['TSS'] = np.where(protein_coding_transcripts['strand'] == '+',
                                             protein_coding_transcripts['start'],
                                             protein_coding_transcripts['end'])
```

```python
#create column with transcript ID without version number; useful for comparing to transcript sets from previous gencode versions
protein_coding_transcripts['transcript_id_without_version'] = protein_coding_transcripts['ID'].str.split('.',expand=True)[0]
```

```python
protein_coding_transcripts.to_csv('../../Data/GENCODE48/gencode_48_protein_coding_transcripts.csv',index=False)
```
