```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from copy import deepcopy 
import gpplot as gpp
gpp.set_aesthetics(context='paper')
```

```python
graph_pathout = '../../Figures/Figure 6/'
```

```python
# import combined dataset
file_path = "../../Data/primaryLibrary/"
merged_df = pd.read_csv(file_path + "inHouseDatasetsCombined.csv")
merged_df.head(1)
```

```python
selectGene = 'ZNF131'
selectGeneDf = merged_df[merged_df['Target Gene Symbol'] == selectGene].copy()
selectGeneDf
```

```python
tss_p1 = 43120915 # from CRISPick
tss_p2 = 43121616 # from UCSC

#custom palette for the scatter plot
palette = {'other': '#1f77b4', 'p1 optimal': 'red', 'p2 optimal': 'green'}

#create new column to define the color categories
selectGeneDf['color'] = 'other'  # default blue
selectGeneDf.loc[selectGeneDf['sgRNA \'Cut\' Site TSS Offset'].between(0, 75), 'color'] = 'p1 optimal'  # red for range p1
selectGeneDf.loc[selectGeneDf['sgRNA \'Cut\' Site TSS Offset'].between(tss_p2-tss_p1, tss_p2-tss_p1+75), 'color'] = 'p2 optimal'  # green for range p2

plt.figure(figsize=(6,4))
ax = sns.scatterplot(data = selectGeneDf, x='sgRNA \'Cut\' Site TSS Offset', 
                y='z_scored_avg_lfc_A549_Kox1',
                   hue='color', palette=palette, legend=False)
sns.despine()
y_min, y_max = -9, 5
ax.set_ylim(y_min, y_max)
plt.fill_betweenx(y=[y_min, y_max], 
                  x1=0, x2=75, color='lightgrey', alpha=0.5)
plt.text(-300, 5, 'MANE select TSS', color='red', ha='center', va='top', fontsize=10)


plt.fill_betweenx(y=[y_min, y_max], 
                  x1=tss_p2-tss_p1, x2=tss_p2-tss_p1+75, color='green', alpha=0.2)
plt.text(475, 4, 'Altenative TSS', color='green', ha='center', va='top', fontsize=10)

plt.title('guides designed to target ZNF131 (essential gene)')
plt.ylabel('Z-Score (A549 Kox1)')

gpp.savefig(graph_pathout + "ZNF131_scatter_alternative_tss_labeled.png", dpi=300, bbox_inches='tight')
```
