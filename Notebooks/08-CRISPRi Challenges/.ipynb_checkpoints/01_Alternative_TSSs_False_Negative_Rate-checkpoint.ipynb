```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import gpplot as gpp
gpp.set_aesthetics(context = 'paper')
```

```python
#read in tiling data
crispri_tiling = pd.read_csv('../../Data/cleanedData/CombinedCleanedDatasets.csv')

crispri_tiling['cell_line'] = crispri_tiling['condition']

#average z-scores across different screens for each guide
crispri_tiling = crispri_tiling.groupby(['Target Gene Symbol',
                                         'sgRNA Context Sequence',
                                         'sgRNA \'Cut\' Position',
                                         'sgRNA \'Cut\' Site TSS Offset',
                                         'TSS Position',
                                         'Strand of Target',
                                         'cell_line',
                                         'rs3ChenSeqScore'])['signed_zscore'].mean().reset_index()

crispri_tiling
```

```python
#merge with gencode 48 transcripts using gene symbol
all_gencode_48_transcripts = pd.read_table('../../Data/Transcript Flags/gencode_48_protein_coding_transcripts.txt')
all_gencode_48_transcripts = all_gencode_48_transcripts[['gene_name','transcript_id','TSS','tag','transcript_type','transcript_support_level',
                                                         'level']].drop_duplicates()
all_gencode_48_transcripts
```

```python
tiling_transcripts_merged = crispri_tiling.merge(all_gencode_48_transcripts,
                                                 left_on = 'Target Gene Symbol',
                                                 right_on = 'gene_name',
                                                 how = 'inner')
tiling_transcripts_merged
```

```python
#label mane vs alternative TSSs
tiling_transcripts_merged['MANE Select'] = tiling_transcripts_merged['tag'].str.contains('MANE_Select') == True
```

```python
#make a column for MANE TSS of each gene
mane_tsss = tiling_transcripts_merged[tiling_transcripts_merged['MANE Select']][['Target Gene Symbol','TSS']].drop_duplicates()
mane_tsss = mane_tsss.rename(columns = {'TSS':'MANE TSS'})
tiling_transcripts_merged = tiling_transcripts_merged.merge(mane_tsss)
tiling_transcripts_merged
```

```python
#calculate tss offset
tiling_transcripts_merged['tss_offset'] = np.where(tiling_transcripts_merged['Strand of Target'] == '+',
                                                   tiling_transcripts_merged['sgRNA \'Cut\' Position'] - tiling_transcripts_merged['TSS'],
                                                   tiling_transcripts_merged['TSS'] - tiling_transcripts_merged['sgRNA \'Cut\' Position'])

#plot tss offset vs z-score as a sanity check
plot_df = tiling_transcripts_merged[(tiling_transcripts_merged['MANE Select'])]
sns.scatterplot(data = plot_df, x = 'tss_offset', y = 'signed_zscore')
```

```python
#define optimal region relative to each TSS
tiling_transcripts_merged['opt_region'] = (tiling_transcripts_merged['tss_offset'] > 0) & (tiling_transcripts_merged['tss_offset'] < 75)
```

```python
#subset to active (high rs3seq and in optimal region) guides
good_guides = tiling_transcripts_merged[tiling_transcripts_merged['opt_region'] & (tiling_transcripts_merged['rs3ChenSeqScore'] > 0)]
```

```python
#calculate average z-score per TSS
average_zscore_df = good_guides.groupby(['Target Gene Symbol',
                                         'cell_line','TSS',
                                         'MANE TSS','MANE Select']).agg(average_zscore = ('signed_zscore','mean'),
                                                                        num_guides = ('sgRNA Context Sequence','nunique')).reset_index()
#limit to those with at least two guides
average_zscore_df = average_zscore_df[average_zscore_df['num_guides'] > 1]
average_zscore_df
```

```python
#subset to alternative TSSs that are sufficiently far from the MANE TSS
average_zscore_df['dist_from_MANE'] = np.abs(average_zscore_df['MANE TSS'] - average_zscore_df['TSS'])
average_zscore_df = average_zscore_df[(average_zscore_df['dist_from_MANE'] > 25) | average_zscore_df['MANE Select']]
average_zscore_df
```

```python
#choosing a cutoff for a tss which "worked" and a tss which didn't
average_zscore_df[average_zscore_df['MANE Select']]['average_zscore'].hist(bins=50)
```

```python
average_zscore_df['average_zscore'].hist(bins=50)
```

```python
average_zscore_df[~average_zscore_df['MANE Select']]['average_zscore'].hist(bins=50)
```

```python
average_zscore_df['worked'] = average_zscore_df['average_zscore'] > 2
average_zscore_df['failed'] = average_zscore_df['average_zscore'] < 2
```

```python
#remove ambiguous cases
plot_df = average_zscore_df[average_zscore_df['worked'] | average_zscore_df['failed']]

#make dfs of mane and alternative tsss that worked
mane_tss_worked_df = plot_df[plot_df['MANE Select']][['Target Gene Symbol','cell_line','worked']].rename(columns = {'worked':'mane_tss_worked'})
alt_tss_worked_df = plot_df[~plot_df['MANE Select']].groupby(['Target Gene Symbol','cell_line']).agg(alt_tss_worked = ('worked','any')).reset_index()

#merge into one df
new_plot_df = mane_tss_worked_df.merge(alt_tss_worked_df,how='inner')

#rename categories
new_plot_df['Category'] = new_plot_df['mane_tss_worked'].astype(str) + '_' + new_plot_df['alt_tss_worked'].astype(str)
new_plot_df[''] = new_plot_df['Category'].map({'True_True':'MANE and Alternative TSS(s) Worked',
                                               'True_False':'MANE TSS Only Worked',
                                               'False_True':'MANE TSS Failed, Alternative TSS(s) Worked',
                                               'False_False':'No TSS Worked'})

#calculate proportion of genes in each category
new_plot_df['total_genes'] = new_plot_df.groupby('cell_line')['Target Gene Symbol'].transform('nunique')
new_plot_df['num_genes'] = new_plot_df.groupby(['cell_line','Category'])['Target Gene Symbol'].transform('nunique')
new_plot_df['prop_genes'] = new_plot_df['num_genes']/new_plot_df['total_genes']
new_plot_df
```

```python
new_plot_df = new_plot_df[['cell_line','','prop_genes','total_genes']].drop_duplicates()
new_plot_df
```

```python
#pivot df to make stacked barplot
pivoted_df = new_plot_df.pivot(index = 'cell_line', columns = '', values = 'prop_genes')
pivoted_df = pivoted_df[['MANE and Alternative TSS(s) Worked','MANE TSS Only Worked','MANE TSS Failed, Alternative TSS(s) Worked','No TSS Worked']]

#plot
pivoted_df.plot(kind='bar',stacked=True)
plt.legend(bbox_to_anchor = [1,1])
sns.despine()
plt.xticks(rotation=0)
plt.xlabel('Cell Line')
plt.ylabel('Proportion of Genes')
plt.title('MANE vs alternative TSS usage in Cas9 CRISPRi tiling data')

#annotate with number of genes
x = 0
for cell_line in pivoted_df.index:
    n = new_plot_df.loc[new_plot_df['cell_line'] == cell_line,'total_genes'].tolist()[0]
    plt.text(x,1.01,str(n),ha='center')
    x += 1

#plt.savefig('../../Figures/mane_vs_alternative_tss_usage_scale_tiling_data_stacked_barplot.pdf',dpi=300,bbox_inches='tight')
```
